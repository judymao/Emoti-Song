```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from time import sleep
import csv
import json
import re
import pandas as pd
import numpy as np
```

```python
# Start with scraping happy songs
happy_songs_df = pd.read_csv("happy_songs.csv", header=None)
happy_titles = np.array(happy_songs_df.iloc[:,0])
happy_artists = np.array(happy_songs_df.iloc[:,1])
```

```python
processed_songs = []

for song in happy_titles:
    numbers_in_brackets_removed = re.sub(r'\(.*\)',"",song)
    processed_song = re.sub(r'\W+', '', numbers_in_brackets_removed).lower()
    processed_song = ''.join([i for i in processed_song if not i.isdigit()])
    processed_songs.append(processed_song)

processed_artists = []
for artist in happy_artists:
    numbers_in_brackets_removed = re.sub(r'\(.*\)',"",artist)
    processed_artist = re.sub(r'\W+', '', numbers_in_brackets_removed).lower()
    processed_artist = ''.join([i for i in processed_artist if not i.isdigit()])
    processed_artists.append(processed_artist)
    
```

```python
# url to scrape the lyrics from
base_url = "https://www.azlyrics.com/lyrics/{}/{}.html"

# file in which the lyrics would be saved
lyrics_file = "happy_lyrics_scraped.txt"

lyrics_not_found_for = []

# delay after each execution of call for not exceeding the requests count and also not to overburden the server
delay = 10

with open(lyrics_file, "w") as file:
    
    for i in range(len(processed_songs)):
        artist = processed_artists[i]
        song = processed_songs[i]
        final_url = base_url.format(artist,song)

        try:
            html_page = urlopen(final_url)
            soup = BeautifulSoup(html_page, 'html.parser')

            html_pointer = soup.find('div', attrs={'class':'ringtone'})
            song_name = html_pointer.find_next('b').contents[0].strip()
            lyrics = html_pointer.find_next('div').text.strip()

            file.write("###"+song_name+"###")
            file.write("\n\n")
            file.write(lyrics)
            file.write("\n\n")
            
            print("Lyrics successfully written to file for : " + song_name)
            
        except:
            print("Lyrics not found for : " + song)
            lyrics_not_found_for.append(song)
            
        finally:
            sleep(delay)
```
